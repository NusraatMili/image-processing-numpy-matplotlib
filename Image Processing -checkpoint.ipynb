```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy.ndimage import gaussian_filter


#read image using matplot library
image = mpimg.imread("cat.jpg")

plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
#convert image to 3D array using numpy
image_arr = np.array(image)


#cropping image using slicing
image_crop = image_arr[82:350, 110:540]

plt.imshow(image_crop)
plt.axis('off')
plt.show()
```

```python
#bluring image using gaussian filter
image_blur = gaussian_filter(image, sigma=4)

plt.imshow(image_blur)
plt.axis('off')
plt.show()
```
